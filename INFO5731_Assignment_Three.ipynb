```python
# Write your code here
import pandas as pd
import re
from nltk.corpus import stopwords
from textblob import TextBlob
stop = stopwords.words('english')
import nltk
nltk.download('punkt')
from nltk.stem import PorterStemmer
ps = PorterStemmer()
from textblob import Word

nltk.download('wordnet')
import itertools
import collections
from nltk import ngrams
from nltk.tokenize import word_tokenize

df=pd.read_csv(r'C:\Users\preet\Downloads\rev.csv')
df.columns
# df=df.iloc[:50]
```

```python
df.dropna(inplace=True)
#Lower Case
df['review_lower_case'] = df['Review'].apply(lambda x: " ".join(x.lower() for x in x.split()))
#Punctuation Removal
df['review_removed_punc'] = df['review_lower_case'].str.replace('[^\w\s]','')
df['review_special_char'] = df['review_removed_punc'].apply(lambda x: ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', charctr) for charctr in x ))
#Stopwords Removal
df['review_after_stop_words'] = df['review_special_char'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
#Spelling Correction
df['review_spell_check'] = df['review_after_stop_words'].apply(lambda x: str(TextBlob(x).correct()))
```

```python
#Tokenization

df['review_Tokenization'] = df['review_spell_check'].apply(lambda x: TextBlob(x).words)

#Stemming

df['review_stemming'] = df['review_Tokenization'].apply(lambda x: " ".join([ps.stem(word) for word in x]))

#Lemmatization

df['review_after_lemmetisation'] = df['review_stemming'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
df.to_csv(r'C:\Users\preet\OneDrive\Desktop\rapport.csv')

a= []
for i in df['review_after_lemmetisation']:
  a.append(word_tokenize(i))
cleaned_data = [x for x in a if x != []]
iterable_data = list(itertools.chain.from_iterable(cleaned_data))
trigrams_from_data = nltk.trigrams(iterable_data)
frequency_distribution_of_trigrams = nltk.FreqDist(trigrams_from_data)

from collections import Counter
bigrams_from_data = nltk.bigrams(iterable_data)
frequency_distribution_of_bigrams = nltk.FreqDist(bigrams_from_data)
bigrams_dict = dict(frequency_distribution_of_bigrams)
for i in bigrams_dict:
  print( str(i) + ':' + str(bigrams_dict[i] / iterable_data.count(i[0])))
```

# **Question 2: Undersand TF-IDF and Document representation**

(40 points). Starting from the documents (all the reviews, or abstracts, or tweets) collected for assignment two, write a python program: 

(1) To build the **documents-terms weights (tf*idf) matrix bold text**.

(2) To rank the documents with respect to query (design a query by yourself, for example, "An Outstanding movie with a haunting performance and best character development") by using **cosine similarity**.

```python
# Write your code here
import math
def tf_idf_calculation(sentence,word):
  size = len(sentence.split(" "))
  term_frequency_value = sentence.count(word)/size
  inverse_df_value = 0
  if(term_frequency_value!=0):
    inverse_df_value = math.log(size)/sentence.count(word)
  else:
    return 0;
  return term_frequency_value*inverse_df_value

sentences = df["review_spell_check"].values.tolist()
tokens_list = set([j for i in sentences for j in i.split(" ")])
tf_idf = pd.DataFrame(tokens_list,columns=["token"])
count = 0
for i in sentences:
  tf_idf[str(count)] = tf_idf["token"].apply(lambda x : tf_idf_calculation(i,x))
  count+=1

tf_idf
```

```python

from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize

sentence_for_query= 'He is also fantastic in the action scenes and that is a huge part of this film and you have to believe in the heroes capabilities and I really did'
Xlist = word_tokenize(sentence_for_query)
stop = stopwords.words('english') 
Xset = {w for w in Xlist if not w in stop}

def compute_vector(Xset, Yset):
  rvector = Xset.union(Yset)
  l1 =[];l2 =[]
  for w in rvector: 
      if w in Xset: l1.append(1)
      else: l1.append(0) 
      if w in Yset: l2.append(1) 
      else: l2.append(0)
  return rvector, l1, l2

def compute_cosine(rvector, l1, l2):
  c = 0
  for i in range(len(rvector)): 
        c+= l1[i]*l2[i] 
  cosine = c / float((sum(l1)*sum(l2))**0.5) 
  return cosine

similarity_list = []
for sentence in df['review_spell_check']:
  Ylist = word_tokenize(sentence)
  Yset = {w for w in Ylist if not w in stop}
  Yset = {w for w in Ylist if not w in stop}
  rvector, l1, l2 = compute_vector(Xset, Yset)
  try:
    similarity = compute_cosine(rvector, l1, l2)
  except ZeroDivisionError:
    similarity = 'None'
  similarity_list.append(similarity)
cosine_dataframe = pd.DataFrame(list(zip(df['Review'],similarity_list)), columns=['Reviews','Cosine Similarity'])
cosine_dataframe
```
